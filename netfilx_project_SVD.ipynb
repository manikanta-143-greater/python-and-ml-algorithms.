```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
netflix_dataset=pd.read_csv("/content/combined_data_1.txt",header=None,names=['cust_id','rating'],usecols=[0,1])
```

```python
movie_count=netflix_dataset.isnull().sum()
```

```python
movie_id=None
movie_num=[]
for cust_id in netflix_dataset["cust_id"]:
  if ':' in cust_id:
    movie_id=int(cust_id.replace(':',""))
  movie_num.append(movie_id)
```

```python
movie_num
```

```python
netflix_dataset['movie_id']=movie_num
```

```python
stars=netflix_dataset.groupby('rating')['rating'].agg(['count'])
stars
```

```python
netflix_dataset = netflix_dataset[netflix_dataset['rating'].notna()]
```

```python
netflix_dataset
```

```python
netflix_dataset.info()
```

```python
netflix_dataset["cust_id"]=netflix_dataset["cust_id"].astype(int)
```

```python
netflix_dataset.info()
```

```python
dataset_movie_summary=netflix_dataset.groupby('movie_id')['rating'].agg(["count"])
dataset_movie_summary
```

```python
movie_benchmark=round(dataset_movie_summary['count'].quantile(0.65),0)
movie_benchmark
```

```python
drop_movie_list=dataset_movie_summary[dataset_movie_summary['count']<movie_benchmark].index
len(drop_movie_list)
```

```python
dataset_cust_summary=netflix_dataset.groupby('cust_id')['rating'].agg(["count"])
cust_benchmark=round(dataset_cust_summary['count'].quantile(0.65),0)
cust_benchmark
```

```python
drop_cust_list=dataset_cust_summary[dataset_cust_summary['count']<cust_benchmark].index
len(drop_cust_list)
```

```python
netflix_dataset=netflix_dataset[~netflix_dataset['movie_id'].isin(drop_movie_list)]
netflix_dataset=netflix_dataset[~netflix_dataset['cust_id'].isin(drop_cust_list)]
print('After the triming, the shape is: {}'.format(netflix_dataset.shape))
```
